```
items = json_data['businesses']['[bounds:128.8308842;35.8938726;128.8647962;35.9085655][query:맛집][sortingOrder:reviewCount]']['items']
items[0]
```

```
itemlist = []
for item in items:
  try:
    name = item['name']
    cat = item['category']
    desc = item['desc']
    x,y = float(item['y']), float(item['x']) # change!
    try:
      add = [item['roadAddr'], item['commonAddr']+" "+item['addr']]
    except:
      add = ["",item['commonAddr']+" "+item['addr']]
    rev_cnt = int(item['blogCafeReviewCount'])
    try:
      tags = item['tags']
    except:
      tags = []
    itemlist.append([name,cat,desc,[x,y],add,rev_cnt,tags])
  except:
    print(item)
  
```
